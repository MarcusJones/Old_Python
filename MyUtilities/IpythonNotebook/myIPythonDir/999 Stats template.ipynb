```python
fig = plt.figure()
fig.set_size_inches(*PLOT_WIDE)
axes = fig.add_axes([0, 0, 1, 1])
axes.grid()
# Plot between -10 and 10 with .001 steps.
x = np.arange(-10, 10, 0.001)

h = list()
names = list()
# NORM
h.append(axes.plot(x, stats.norm.pdf(x=x,loc=0,scale=1)))
names.append("Norm1")
h.append(axes.plot(x, stats.norm.pdf(x=x,loc=0,scale=2)))
names.append("Norm2")
# F
[dfn, dfd ] = [5,] * 2 # f has 2 
rv = stats.f(dfn, dfd)
numargs = stats.f.numargs
h.append(axes.plot(x, rv.pdf(x)))
names.append("F")
# chi
[df] = [1/1000,] * 1 # chi2 has 1
rv = stats.chi(df,loc=2,scale=3)
h.append(axes.plot(x, rv.pdf(x)))
names.append("chi")
# chi2
[df] = [1/1000,] * 1 # chi2 has 1
rv = stats.chi2(df,loc=3,scale=3)
h.append(axes.plot(x, rv.pdf(x)))
names.append("chi2")
# t
[df] = [0.9,] * 1 # t has 1
rv = stats.t(df,loc=0,scale=2)
h.append(axes.plot(x, rv.pdf(x)))
names.append("t")
l=axes.legend(names)
```
