```python
def plot_df(df, this_size = (10,5), this_dpi = 400):
    with mpl.rc_context(rc = rc):
        fig = plt.figure(figsize=this_size, dpi=this_dpi)
        axes = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # left, bottom, width, height (range 0 to 1)
        axes.set_xlabel('x')
        axes.set_ylabel('y')
        axes.set_title('title')
        df.plot(ax = axes)
```

```python
def plot_masked(df, mask, this_size = (10,5), this_dpi = 400):
    with mpl.rc_context(rc = rc):
        fig = plt.figure(figsize=this_size, dpi=this_dpi)
        axes = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # left, bottom, width, height (range 0 to 1)
        axes.set_xlabel('x')
        axes.set_ylabel('y')
        axes.set_title('title')
        df.iloc[:,mask].plot(ax = axes)    
```
