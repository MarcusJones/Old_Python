```python
from pylab import *
```

```python
import matplotlib.pyplot as plt
```

```python
x = linspace(0, 5, 10)
y = x ** 2
```

```python
figure()
plot(x, y, 'r')
xlabel('x')
ylabel('y')
title('title')
show()
```

```python
%pylab qt
```

```python
subplot(1,2,1)
plot(x, y, 'r--')
subplot(1,2,2)
plot(y, x, 'g*-');
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # lower, bottom, width, height (range 0 to 1)
axes.plot(x, y, 'r'), axes.set_xlabel('x'), axes.set_ylabel('y'), axes.set_title('title');
```
