```python
def getUpperTriAsVec(matrix):
    """Return upper tri as a vector"""
    assert type(matrix) == numpy.ndarray
    assert matrix.shape[0] == matrix.shape[1]
    upper = np.triu_indices(len(matrix),1)
    upperTri = matrix[upper[0],upper[1]]
    print "Getting upper tri, length {}".format(len(upperTri))
    return upperTri

def plotCorr(A,B,myText,rootPath):
    print "Dataframe A: {}, Dataframe B: {}".format(A.shape, B.shape)
    # 2 dataframes
    # First, get the pairwise correlations
    Acorr = A.corr()
    Bcorr = B.corr()
    print "A correlation size; {}, B correlation matrix size; {}".format(Acorr.shape, Bcorr.shape)
    
    
    # Extract the upper tri as a vector
    A_corrVec = getUpperTriAsVec(Acorr.values)
    B_corrVec =  getUpperTriAsVec(Bcorr.values)
    
    fig = plt.figure()
    fig.set_size_inches(*PLOT_A6)
    axes = fig.add_subplot(111)
    axes.grid()
    
    #axes.set
    axes.set_xlabel("Sleep Correlation",fontsize=12)
    axes.set_ylabel("Wake Correlation",fontsize=12)
    axes.set_title("{}, correlation = {:0.3}".format(myText,numpy.corrcoef(A_corrVec, B_corrVec)[ 0, 1 ]))
    axes.set_xlim([-0.2,0.5])
    axes.set_ylim([-0.2,0.5])
    axes.scatter(A_corrVec,B_corrVec,s=20,facecolor="tomato",marker ='o',lw = 1,edgecolor='b')
    fig.savefig(rootPath+myText+".pdf")


fileName = r"C:\Dropbox\04 School\IST Statistics\HW\p03\TEX\fig\\"
plotCorr(sleep,walk,"Full set",fileName)

#print sleep[0:len(sleep)/2]
fig = plotCorr(sleep,walk[:len(walk)/2],"First half",fileName)
fig = plotCorr(sleep,walk[len(walk)/2:],"Second half",fileName)
```
