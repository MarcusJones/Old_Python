<p>Null Hypothesis $H_{0}: \mu_{ALL} = \mu_{AML}$ $\;\;$ at $\alpha = 0.05$ </p>
<p>Testing for equality by two-tailed two-sample t-test.</p>

```python
# Sample gene, as a trial only
nGene = 1000
ALL1000 = samplesALL[nGene]
AML1000 = samplesAML[nGene]
result = stats.ttest_ind(ALL1000,AML1000)
print "t-value: {:.2f}, p-value: {:.2f}".format(float( result[0] ), float(result[1]))
```

```python
# Helper function
def rejectNullHypothesis(sample1, sample2, target_pVal):
    result = stats.ttest_ind(sample1,sample2)
    result_tVal=float(result[0])
    result_pVal=float(result[1])
    if result_pVal < target_pVal:
        return True
    else:
        return False
assert(rejectNullHypothesis(ALL1000, AML1000,0.05)==False) #Confirmed
```

```python
comparisonSet = zip(samplesALL,samplesAML)
assert(rejectNullHypothesis(*comparisonSet[1000],target_pVal=0.05)==False) # Again confirmed
default_Pval = 0.05
positives = [rejectNullHypothesis(*pair, target_pVal=default_Pval) for pair in comparisonSet]
out("i/ For p-0.05, <i>{}</i> genes are associated".format(sum(positives)))
```

```python
# Benjamini / Hochberg FDR
# 1) First, get the entire t-test results set;
resultSet = list()
for pair in comparisonSet:
    result = stats.ttest_ind(*pair)
    resultSet.append((pair[0],pair[1], result[0], result[1]))
resultFrame = pd.DataFrame(resultSet,columns=["ALL","AML",'t', 'p']).sort("p", ascending=True).reset_index()
resultFrame = resultFrame.sort("p", ascending=False)

# 2) Iterate over the set
significant = list()
correctedList = list()
for row in resultFrame.iterrows():
    rank = row[0] + 1
    thisPval =row[1]["p"]
    correctedP = thisPval # Default
    if thisPval < default_Pval:
        correctedP = len(resultFrame)/rank * thisPval
        #print rank, thisPval
        if correctedP < default_Pval:
            significant.append(True)
            correctedList.append(correctedP)
            continue
    significant.append(False)
    correctedList.append(correctedP)

resultFrame["sig"] = pd.Series(significant, index=resultFrame.index)
resultFrame["corrected"] = pd.Series(correctedList, index=resultFrame.index)
#print resultFrame.ix[[1000],["sig","p","corrected"]]
out( "ii/ After the Benjamini / Hochberg correction, <i>{}</i> genes are associated".format(sum(resultFrame["sig"])))
```

```python
# Bonferroni correction
bonferroniP = default_Pval / len(comparisonSet)
BonferroniPositives = [rejectNullHypothesis(*pair, target_pVal=bonferroniP) for pair in comparisonSet]
out("iii/ For the Bonferroni corrected p-{:0.10f}, <i>{}</i> genes are associated".format(bonferroniP,sum(BonferroniPositives)))
```

```python
def is_outlier(points, thresh=3):
    """
    Returns a boolean array with True if points are outliers and False 
    otherwise.

    Parameters:
    -----------
        points : An numobservations by numdimensions array of observations
        thresh : The modified z-score to use as a threshold. Observations with
            a modified z-score (based on the median absolute deviation) greater
            than this value will be classified as outliers.

    Returns:
    --------
        mask : A numobservations-length boolean array.

    References:
    ----------
        Boris Iglewicz and David Hoaglin (1993), "Volume 16: How to Detect and
        Handle Outliers", The ASQC Basic References in Quality Control:
        Statistical Techniques, Edward F. Mykytka, Ph.D., Editor. 
    """
    if len(points.shape) == 1:
        points = points[:,None]
    median = np.median(points, axis=0)
    diff = np.sum((points - median)**2, axis=-1)
    diff = np.sqrt(diff)
    med_abs_deviation = np.median(diff)

    modified_z_score = 0.6745 * diff / med_abs_deviation

    return modified_z_score > thresh
```
