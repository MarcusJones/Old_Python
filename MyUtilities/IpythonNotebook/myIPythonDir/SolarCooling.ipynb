```python
# Two helper functions to handle times
def trntime2datetime(trn_time):
    matlab_datenum = dt.datetime(2014,1,1,0).toordinal() + 1 + trn_time / 24
    day = dt.datetime.fromordinal(int(matlab_datenum))
    dayfrac = dt.timedelta(days=matlab_datenum%1) - dt.timedelta(days = 366)
    return roundTime(day + dayfrac,60)

def roundTime(this_time=None, roundTo=30):
   """Round a datetime object to any time laps in seconds
   dt : datetime.datetime object, default now.
   roundTo : Closest number of seconds to round to, default 1 minute.
   Author: Thierry Husson 2012 - Use it as you want but don't blame me.
   """
   seconds = (this_time - this_time.min).seconds
   # // is a floor division, not a comment on following line:
   rounding = (seconds+roundTo/2) // roundTo * roundTo
   return this_time + datetime.timedelta(0,rounding-seconds,-this_time.microsecond)
```
