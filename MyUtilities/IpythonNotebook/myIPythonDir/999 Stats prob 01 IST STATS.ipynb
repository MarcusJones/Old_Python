### b) Simulate $\bar{X}$ for various sample sizes;

```python
nMeanSamples = 1000
nTimes = [5, 10, 200]

def getMean(nSamples,vals):
    # Helper function, samples the distribution nSamples times, returns mean
    return np.mean([np.random.choice(vals) for pos in range(nSamples)])

meanSamples = list()
for n in nTimes:
    thisMeanSample = [getMean(n, Xvals) for i in range(nMeanSamples)]
    meanSamples.append([n, thisMeanSample])

for result in meanSamples:
    n, sample = result
    
    fig = sm.graphics.qqplot(np.array(sample), fit=True)
    fig.set_size_inches(PLOTSIZE,PLOTSIZE)
    axis = fig.axes[0]
    axis.set_ylim([-5,5])
    axis.set_xlim([-5,5])
    axis.set_title('QQ plot against norm, mean(X) sampled n = {} times'.format(n));
    fig.show()
```

<b>Discussion 1.c):</b> The central limit theorem is therefore illustrated, a sample $\bar{X}$, approaches the normal distribution for large sample size.

## Question 2: 

<p>Sample = 100 independent values of a norm. </p>
<p>Let $X$ be a normally distributed variable with $\mu = 0$</p>
<p>Null Hypothesis $H_{0}: \mu = 0.0$ at $\alpha = 0.05$ </p>
<p>$\alpha$ is the probability of a the Null-hypothesis not being rejected when it should be, i.e. a false-positive.
<p>We know that the Null Hypothesis should NOT be rejected.</p>
<p>Testing for equality is a two-tailed t-test.</p>

```python
#### (DEVELOPMENT, IGNORE) ####
# Using scipy;
n = 100
sample = numpy.random.standard_normal(n)
alpha = 0.05
popmean = 0.0
t, prob = stats.ttest_1samp(sample, popmean)
print "t-value:", t
print "p-value", prob # Will reject Ho when prob < alpha
```

```python
#### (DEVELOPMENT, IGNORE) ####
# Validation of t, manual calculation
muHo = 0.0
muSample = mean(sample)
diff = muSample - muHo
stDev = std(sample)
stdErr = stats.sem(sample)
tStat = diff / stdErr
print "t-value:", tStat
```

#### 2.a): Now to simulate this null-hypothesis and t-test;

```python
# First set a helper function
def nullHypothesisTrue(distrib, n, alpha, popMean):
    sample = distrib(n) # Get a new sample
    t, prob = stats.ttest_1samp(sample, popmean)
    if prob <= alpha:
        return False
    else:
        return True
# Now simulate
n = 100
distrib = numpy.random.standard_normal
popmean = 0.0
m = 10000
result = [nullHypothesisTrue(distrib, n, alpha, popmean) for i in range(m)]
print "Over {} simulations of sample size {}, Ho is rejected (falsely) {} times ({:0.2f}%)".format(m,n,result.count(False), result.count(False)/m*100)
#print
#print myNullHypothesis(n,alpha,popmean)
```

<b>Discussion 2.a):</b>   The 95% confidence interval of the t-test is therefore demonstrated.

```python
n = 10
distrib = numpy.random.standard_normal
popmean = 0.0
m = 10000
result = [nullHypothesisTrue(distrib, n, alpha, popmean) for i in range(m)]
print "Over {} simulations of sample size {}, Ho is rejected (falsely) {} times ({:0.2f}%)".format(m,n,result.count(False), result.count(False)/m*100)
```

<b>Discussion 2.b):</b> The robustness of the t-test even for small sample sizes is demonstrated, when the underlying population variable is normally distributed.

```python
n = 10
sigma = 10
mu = 31
degreesFreedom = n - 1
p = 0.01
# Need critical t two-valued for 99%, p <0.01
# isf is one-tailed, so need to take p/2!
zScore = stats.norm.isf(p/2, degreesFreedom, loc=0, scale=1)
print "z-score = {:.3}".format(zScore)
givenCI = zScore * sigma / sqrt(n)
print "Confidence interval total width = {:.3}ppb".format(2*givenCI)
print "{:.3} < mu < {:.3}".format(mu - givenCI,mu + givenCI)
```
